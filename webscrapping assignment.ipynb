# importing libraries

```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

# scraping website

```python
url ='https://www.flipkart.com/search?q=jewellery&sid=mcr%2C96v%2Cyx2&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=jewellery%7CJewellery+Sets&requestId=21b662dd-0f47-44ac-921f-7b9cc313609f&as-searchtext=jew'
```

# website response

```python
res=requests.get(url)
```

```python
res
```


# parsing

```python
soup = BeautifulSoup(res.text)
```

```python
soup
```

```python
jewellery_name= soup.select('._2WkVRV')
```

```python
jewellery_name
```

```python
price_name=soup.select('._3bPFwb ._30jeq3')
```

```python
price_name
```

```python
offer_name=soup.select('div+ div ._3bPFwb span')
```

```python
offer_name
```

```python
df=pd.DataFrame( jewellery_name,columns=['jewellery_names'] )
```

```python
df
```

```python
df['offer_name']=offer_name
```

```python
df
```

```python
df.to_csv("ddetails.csv")
```
